### corr(상관관계) 함수 이용하기

### 변수(column) 사이의 상관계수(correlation)
- corr 함수를 통해 상관계수 연산 (-1,1 사이의 결과)
  - 연속성(숫자형) 데이터에 대해서만 연산
  - 인과관계를 의미하진 않음
- 두 변수간의 흐름이 어떻게 이루어지고 있는지에 대한 관계 연산
- **1에 가까울수록 비슷하게 증가한다. -1이라면 그 반대**
- **대각선 값은 항상 1, 대칭되는 원소값들은 같다는 특징을 가진다**

```python
train_data.corr()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.matshow(train_data.corr())
#색이 밝을수록 상관관계가 깊다는 것을 의미함
```

### 데이터는 크게 숫자형 데이터와 범주형 데이터가 있음

#### 숫자형 데이터(Numerical Type)
- 연속성을 띄는 숫자로 이루어진 데이터
  - ex) Age, Fare 등

#### 범주형 데이터(Categorical Type)
- 연속적이지 않은 값(대부분의 경우 숫자를 제외한 나머지 값)을 갖는 데이터
  - ex) Name, Sex, Ticket 등
- 어떤 경우, 숫자형 타입이라 할지라도 개념적으로 범주형으로 처리해야 할 경우가 있음
  - ex) pclass
- 머신러닝 데이터에서 바로 계산이 불가능하다는 특징을 가진다.

## 숫자 데이터를 범주형 데이터로 바꾸어보자

#### Pclass 변수 변환해보기
- astype 사용하여 간단히 타입만 변환

```python
train_data['Pclass']=train_data['Pclass'].astype(str)
```

```python
train_data.info() #pclass가 바뀐 것을 확인
```

#### Age 변수 변환해보기
- 변환 로직을 함수로 만든 후, apply 함수 적용
- numpy와 pandas에서는 **loop을 최대한 지양하여야 한다.**

```python
import math
def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age/10)*10
```

```python
train_data['Age'].apply(age_categorize)
```

## 이제 범주형 데이터를 숫자형 데이터로 바꾸어보자(전처리)

### one-hot encoding
- 범주형 데이터는 분석단계에서 계산이 어렵기 때문에 숫자형으로 변경이 필요함
- 범주형 데이터의 각 범주(category)를 column 레벨로 변경
- 해당 범주에 해당하면 1, 아니면 0으로 채우는 인코딩 기법
- pandas.get_dummies 함수 사용
  - drop_first : 첫번째 카테고리 값은 사용하지 않음
    - 각 속성에 대해 1은 하나만 존재하기 때문에 하나 없애도 값을 유추해낼 수 있음

```python
pd.get_dummies(train_data) #프로그램이 판단하고, 모든 속성에 대해 컬럼 단계로 높인다.
```

```python
pd.get_dummies(train_data,columns=['Pclass','Sex','Embarked']) 
#지정한 속성값에 대해서만 one-hot encoding이 적용됨
```

```python
pd.get_dummies(train_data,columns=['Pclass','Sex','Embarked'],drop_first=True) 
#첫번째 속성이 없어져도 우리는 충분히 값을 유추해낼 수 있음
```
