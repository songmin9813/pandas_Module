### group by
- 아래의 세 단계를 적용하여 데이터를 그룹화(groupping)
  - 데이터 분할
  - operation 적용
  - 데이터 병합
- SQL의 group by와 개념적으로 동일. 사용법 유사

```python
df=pd.read_csv('C:/Users/mitha/OneDrive/바탕 화면/train.csv')
df.head()
```

### GroupBy groups 속성
- 각 그룹과 그룹에 속한 index를 dict 형태로 표현

```python
class_group=df.groupby('Pclass')#리스트를 인자로 넘겨받아 그룹핑
class_group
```

```python
class_group.groups #같은 속성끼리 그룹화 시킨 것을 확인할 수 있다.
```

```python
gender_group=df.groupby('Sex')
gender_group.groups#남자와 여자 두 개이기 때문에 두 개의 그룹을 가지고 있을 것
```

```python
class_group.count() #NaN 값은 빠짐
```

```python
#성별에 따른 생존율 구하기
df.groupby('Sex').mean()['Survived']
```

### 복수 columns로 groupping 하기
- groupby에 columns 리스트 전달
- 통계함수를 적용한 결과는 multiindex를 갖는 dataframe

```python
df.groupby(['Pclass','Sex']).mean()
```

```python
df.groupby(['Pclass','Sex']).mean().index
#멀티 인덱스를 가지는 것을 알 수 있다.
```

```python
#접근하기 위해서 다음과 같이 사용
df.groupby(['Pclass','Sex']).mean().loc[(2,'female')]
#튜플 형태로 인덱스를 받아서 접근.
```

### index를 이용한 group by
- index가 있는 경우, groupby 함수에 level 사용 가능
  - level은 index의 depth를 의미하며, 가장 왼쪽부터 0부터 증가
  
  
- set_index 함수
  - column 데이터를 index 레벨로 변경
- reset_index 함수
  - 인덱스 초기화

```python
df.set_index('Pclass')#피클래스가 인덱스로 바뀐
```

```python
df.set_index(['Pclass','Sex'])#여러 인덱스로도 사용이 가능
```

```python
df.set_index(['Pclass','Sex']).reset_index()#이걸로 모든 인덱스 초기화 후 0-based로 돌아감
```

```python
df.set_index('Age').groupby('Age').mean()
```

```python
df.set_index('Age').groupby(level=0).mean()
#여러 인덱스를 가지고 있다면 level을 통해서 몇 번째 인덱스를 지정할 것인지도 선택이 가능
#위의 코드는 바로 위의 'Age'를 지정하는 코드와 기능상 동일하다.
```

### 나이대별로 생존율 구해보기

```python
import math
def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age/10)*10
```

```python
df.set_index('Age').groupby(age_categorize).mean()['Survived']
```

#### MultiIndex를 이용한 groupping

```python
df.set_index(['Pclass','Sex']).groupby(level=[0,1]).mean()
```

#### aggregate(집계) 함수 사용하기
- groupby 결과에 집계함수를 적용하여 그룹별 데이터 확인 가능

```python
df.set_index(['Pclass','Sex']).groupby(level=[0,1]).aggregate([np.mean,np.sum,np.max])
#각 컬럼의 지정한 함수를 집계하는 것을 확인할 수 있다.
```

### transform 함수
- groupby 후 transform 함수를 사용하면 원래의 index를 유지한 채로 통계함수를 적용한다.
- 전체 데이터의 집계가 아닌 각 그룹에서의 집계를 계산
- **따라서 새로 생성된 데이터를 원본 dataframe과 합치기 쉬움**

```python
df.groupby('Pclass').mean()
#인덱스가 줄어드는 것을 확인할 수 있음
```

```python
df.groupby('Pclass').transform(np.mean)
#이러면 축약된 형태가 아닌 전체 컬럼의 형태로 화면에 출력이 됨
```

### pivot
- dataframe의 형태를 변경
- 인덱스, 컬럼, 데이터로 사용할 컬럼을 명시

```python
df = pd.DataFrame({
    '지역': ['서울', '서울', '서울', '경기', '경기', '부산', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일': ['월요일', '화요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
                  })
df
```

### pivot_table
- 기능적으로 pivot과 동일
- pivot과의 차이점
  - 중복되는 모호한 값이 있을 경우, aggregation 함수 사용하여 값을 채운다

```python
pd.pivot_table(df,index='요일',columns='지역',aggfunc=np.mean)
```

### stack & unstack
- stack : 컬럼 레벨에서 인덱스 레벨로 dataframe 변경
  - 즉, 데이터를 쌓아올리는 개념으로 이해하면 쉬움
- unstack : 인덱스 레벨에서 컬럼 레벨로 dataframe 변경
  - stack의 반대 연산
- 둘은 역의 관계에 있음
- level을 명시하지 않으면 기본적으로 -1(맨 뒤)

```python
new_df=df.set_index(['지역','요일'])
new_df
```

```python
new_df.unstack(0)#0번째 레벨(지역)이 컬럼으로 올라감
```

```python
new_df.unstack(1)#1번째 레벨(요일)이 컬럼으로 올라감
```

```python
new_df.unstack(0).stack(0)
```

```python
new_df.unstack(0).stack(1)
```

### dataframe merge
- SQL의 join 처럼 **특정한 column을 기준으로 병합**
  - join 방식 : how 파라미터를 통해 명시
    - inner : 기본값, 일치하는 값이 있는 경우
    - left : left outer join
      - 왼쪽을 기준으로 오른쪽을 채운다. 왼쪽에만 데이터가 있다면 오른쪽은 NaN로 채워짐
    - right : right outer join
    - outer : full outer join
  - pandas.merger 함수가 사용됨

```python
customer = pd.DataFrame({'customer_id' : np.arange(6), 
                    'name' : ['철수'"", '영희', '길동', '영수', '수민', '동건'], 
                    '나이' : [40, 20, 21, 30, 31, 18]})

customer
```

```python
orders = pd.DataFrame({'customer_id' : [1, 1, 2, 2, 2, 3, 3, 1, 4, 9], 
                    'item' : ['치약', '칫솔', '이어폰', '헤드셋', '수건', '생수', '수건', '치약', '생수', '케이스'], 
                    'quantity' : [1, 2, 1, 1, 3, 2, 2, 3, 2, 1]})
orders.head()
```

- on
  - join 대상이 되는 column 명시

```python
pd.merge(customer,orders,on='customer_id')
#기본 방법이 inner이기 때문에 없는 값은 출력하지 않는다.
```

```python
pd.merge(customer,orders,on='customer_id',how='left')
```

```python
pd.merge(customer,orders,on='customer_id',how='right')
```

```python
pd.merge(customer,orders,on='customer_id',how='outer')
#left와 right가 전부 채워진 형태
```

- index 기준으로 join 하기

```python
cust1=customer.set_index('customer_id')
order1=orders.set_index('customer_id')
```

```python
cust1
```

```python
order1
```

```python
pd.merge(cust1,order1,left_index=True,right_index=True)
```
